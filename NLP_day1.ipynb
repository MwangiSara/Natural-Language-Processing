```python
from nltk.tokenize import PunktSentenceTokenizer # it splits  a text where there is punctuation (. ? !)
# tokenize a sentence, word
my_text = "this is an example of a text. It has multiple sentences. here is another sentence. how many sentences are in this text?"
```

```python
# pass our sentence in the tokenizer
tokenizer = PunktSentenceTokenizer()
sentences = tokenizer.tokenize(my_text)
sentences
```

```python
for sentence in sentences:
    print (sentence)
```

```python
# fetching data from the website, the lib used are: requests, bs4(BeautifulSoup)
import requests  # this module helps to send HTTP requests(get,post,delete,put) in python. it is useful in this case for wed scrapping
from bs4 import BeautifulSoup # bs4 is used for pulling data out of a HTML or XML file. beautifulsoup is used to make it easy to scrape  info from web pages
URL = "https://www.standardmedia.co.ke/sports/football/article/2001496512/mbappe-vinicius-bellingham-to-form-real-madrids-dream-attacking-trio"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser") # it parses the downloaded HTML content using the HTML.parser. THis creates a BeautifulSoap onbject called soup with represents the webpage structure
div = soup.findAll("div", class_ ="story") 
soup = BeautifulSoup(str(div)) # creates a new Beautifulsoup object from string reprsentation of the article in div with class story
text = soup.get_text() # get_text extracts all the text content from within the new soup variable/object
print(text)
```

```python
# nltk.download("punkt")
from nltk.tokenize import sent_tokenize, word_tokenize 
# sent_tokenize splits article into sentences, word_tokenize splits article into words
token_sentences = sent_tokenize(text)
token_sentences
token_words = word_tokenize(text)
token_words
```

```python
# nltk.download("stopwords") # stopwords are commoly used words(the,a,an)
from nltk.corpus import stopwords 
print(stopwords.fileids()) # it displays the list of idetentifiers available stopwords set in different languages(english)
mystopwords = set(stopwords.words('english')) # creating a set of commont english stopwords from corpus inst
mystopwords
```

```python
#filter the stopwords from the article
filtered_words=[]
#for every word in token_words:
for word in token_words:
    # if word is in mystopwords:
    if word.lower() not in mystopwords:
        filtered_words.append(word) 
print("the new words without the common words is:",filtered_words)
```

```python
# remove any words from filterd_words that contain non-alphabetic charactes eg punctuation marks, symbols
# for word in filtered_words:
#     # print(word)
#     if word.isalpha():
#         clean_words = word
#         print(clean_words)
clean_words = [word for word in filtered_words if word.isalpha()]
clean_words
```

```python
#analyze the most frequent words in our cleaned words
from nltk.probability import FreqDist
frequency = FreqDist(clean_words)
frequency.most_common(30)
```

```python
#  generating a word cloud image based on the text
from wordcloud import WordCloud, STOPWORDS  #This line imports the WordCloud class and the built-in stopwords list from the wordcloud library. WordCloud is used to generate the visual representation of words, and STOPWORDS provides a list of common words (like "the", "a", "an") that are often excluded from word clouds.

from PIL import Image #This line imports the Image class from the Pillow Fork (PIL) library, which is commonly used for image processing in Python. This allows you to open and display the generated word cloud image.

#Function to Create Wordcloud
def create_wordcloud(text):
        stopwords = set(STOPWORDS) #creates a set containing the stopwords from the imported STOPWORDS list
        wc = WordCloud(background_color="white", #Sets the background color of the word cloud to white.
        max_words=3000, #Limits the number of words displayed in the word cloud to 3000 (you can adjust this value).
        stopwords=stopwords,repeat=True) #Specifies the list of stopwords to exclude from the word cloud (uses the set you created earlier). repeat=True Allows words to appear multiple times in the word cloud, scaled by their frequency (optional)
        wc.generate(str(text)) # This method analyzes the text and generates the word cloud image based on word frequencies.
        wc.to_file("wc.png") #to_file method saves the generated word cloud as a PNG image file named "wc.png"
        print("Word Cloud Saved Successfully") # This line prints a message indicating successful word cloud generation.
        path="wc.png" 
        display(Image.open(path))  #print the image
```

```python
create_wordcloud(frequency.most_common(30))
```
